```python
import os
from dotenv import load_dotenv


load_dotenv("DB_details.env")  

db_config = {
    "host": os.getenv("DB_HOST"),
    "port": int(os.getenv("DB_PORT")),
    "user": os.getenv("DB_USER"),
    "password": os.getenv("DB_PASS"),
    "database": os.getenv("DB_NAME")

}


print("Loaded Database config successfully", db_config)
```

```python
import mysql.connector

def find_customer_by_phoneNum(phone):
    try:
        conn = mysql.connector.connect(**db_config)
        cur = conn.cursor()
        cur.execute("""
            SELECT custId, fname, sname, address, phone
            FROM customers
            WHERE phone = %s
        """, (phone,))
        row = cur.fetchone()
        cur.close()
        conn.close()
        return row
    except Exception as err:
        print("Error (find):", err)
        return None


def register_customer(fname, sname, address, phone):
    try:
        
        if find_customer_by_phoneNum(phone):
            print("This customer already exists.")
            return False

        conn = mysql.connector.connect(**db_config)
        cur = conn.cursor()
        cur.execute("""
            INSERT INTO customers (fname, sname, address, phone)
            VALUES (%s, %s, %s, %s)
        """, (fname, sname, address, phone))
        conn.commit()
        cur.close()
        conn.close()
        print("Customer has been registered successfully.")
        return True
    except Exception as err:
        print("Error (register):", err)
        return False
```
