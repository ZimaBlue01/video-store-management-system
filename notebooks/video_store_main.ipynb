```python
def prompt_int(msg):
    s = input(msg).strip()
    try:
        return int(s)
    except ValueError:
        print("Please enter a valid number.")
        return None


def menu():
    while True:
        print("\n=== Video Store ===")
        print("======================")
        print("1) Register Customer")
        print("======================")
        print("2) Register Movie")
        print("======================")
        print("3) Hire Out Movie")
        print("======================")
        print("4) Return Movie")
        print("======================")
        print("5) Check Movie Availability")
        print("======================")
        print("0) Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            fname = input("First name: ").strip()
            sname = input("Surname: ").strip()
            address = input("Address: ").strip()
            phone = input("Phone: ").strip()
            register_customer(fname, sname, address, phone)

        elif choice == "2":
            vname = input("Movie name: ").strip()
            vtype = input("Type (R=new / B=old): ").strip()
            movie_registration(vname, vtype)

        elif choice == "3":
            phone = input("Customer phone: ").strip()
            cust = find_customer_by_phoneNum(phone)
            if not cust:
                print("Customer not found.")
                continue
            custId = cust[0]  # (custId, fname, sname, address, phone)

            videoId = prompt_int("Video ID to hire: ")
            if videoId is None:
                continue
            videoVer = prompt_int("Video version: ")
            if videoVer is None:
                continue

            movie_hire(custId, videoId, videoVer)

        elif choice == "4":
            videoId = prompt_int("Video ID to return: ")
            if videoId is None:
                continue
            videoVer = prompt_int("Video version to return: ")
            if videoVer is None:
                continue
            movie_returns(videoId, videoVer)

        elif choice == "5":
            videoId = prompt_int("Video ID: ")
            if videoId is None:
                continue
            videoVer = prompt_int("Video version: ")
            if videoVer is None:
                continue
            movie_availibility(videoId, videoVer)

        elif choice == "0":
            print("Goodbye!")
            break
        else:
            print("Invalid option. Try again.")

if __name__ == "__main__":
    menu()
```
